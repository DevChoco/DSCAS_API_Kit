```python
# 수집한 무신사 이미지 전처리 코드
# 1. jpg, webp -> png 변경
# 2. 크기 조정 (보편적 216*216, 512*512 사용)
# 경로와 크기는 코드상 주석을 보고 수정해서 사용해주세요.

from PIL import Image
import os

# 1. 디렉토리 설정 (경로 재설정 가능)
input_dir = "C:/Users/hj123/OneDrive/바탕 화면/전처리전"  # 입력 폴더ㄴ 경로 (직접입력!!!)
output_dir = "C:/Users/hj123/OneDrive/바탕 화면/변환완료(원본)"  # 출력 폴더 경로 (직접입력!!!)

# 출력 폴더가 없으면 생성
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 2. 전처리 함수
def preprocess_image(input_path, output_path, size=(768, 1024)): # 변환하고자 하는 해상도 직접 입력!!!
    with Image.open(input_path) as img:
        # RGB로 변환
        img = img.convert('RGB')
        # 크기 조정 (고품질 보간법)
        img_resized = img.resize(size, Image.Resampling.LANCZOS)
        # PNG로 저장 (무손실)
        img_resized.save(output_path, 'PNG')    

# 3. 모든 .webp와 .jpg 파일 처리
print(f"{input_dir}에서 .webp와 .jpg 파일을 순차적으로 처리합니다.")
for filename in os.listdir(input_dir):
    if filename.endswith(('.webp', '.jpg')):
        input_path = os.path.join(input_dir, filename)
        output_filename = filename.replace('.webp', '.png').replace('.jpg', '.png')
        output_path = os.path.join(output_dir, output_filename)
        preprocess_image(input_path, output_path, size=(768, 1024))  # 변환하고자 하는 해상도 직접 입력!!!
        print(f"변환 완료: {output_filename}")

# 4. 결과 확인
print("전처리 완료! 변환된 파일은 '변환완료(원본)' 폴더에 저장되었습니다.")
print("변환된 파일 목록:")
for file in os.listdir(output_dir):
    print(file)
```
